```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='ProductCategoryKey', y='Category', data=df4)
# plt.title("Violin Plot of Categories and Numerical Values")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x='ProductCategoryKey', y='Category', data=df4)
plt.title("Swarm Plot of Categories and Numerical Values")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.stripplot(x='ProductCategoryKey', y='Category', data=df4, jitter=True)
plt.title("Strip Plot of Categories and Numerical Values")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.pointplot(x='ProductCategoryKey', y='Category', data=df4, ci="sd", capsize=0.1)
plt.title("Point Plot of Categories and Numerical Values with Confidence Intervals")
plt.show()
```

```python
sns.catplot(x='ProductCategoryKey', y='Category', data=df4, kind="violin", height=6, aspect=1.2)
plt.title("Catplot Violin of Categories and Numerical Values")
plt.show()
```
